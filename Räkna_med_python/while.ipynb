```python
'''
    1L mj√∂lk 1500000 bakterier i rumstemp
    Bakterier √∂kar med 50% per timme
    Mj√∂lk sur n√§r bakterier mer √§n 10 000 000
    Antal timmar tils mj√∂lken blir sur?
'''

import matplotlib.pyplot as plt 
import numpy as np

bakterier = 1.5e6
f√∂r√§ndringsfaktor = 1.5
h = 0

y = [bakterier]

while bakterier < 1e7:
    bakterier *= f√∂r√§ndringsfaktor
    h += 1
    y.append(bakterier)

print(h)

x = np.arange(h+1)
print(x)
plt.plot(x,y, '*-')
plt.xlabel("Tid i timmar")
plt.ylabel("Antal bakterier")
plt.title("Bakterier i mj√∂lk i rumstemperatur")


```
