```python
'''
    1L mjölk 1500000 bakterier i rumstemp
    Bakterier ökar med 50% per timme
    Mjölk sur när bakterier mer än 10 000 000
    Antal timmar tils mjölken blir sur?
'''

import matplotlib.pyplot as plt 
import numpy as np

bakterier = 1.5e6
förändringsfaktor = 1.5
h = 0

y = [bakterier]

while bakterier < 1e7:
    bakterier *= förändringsfaktor
    h += 1
    y.append(bakterier)

print(h)

x = np.arange(h+1)
print(x)
plt.plot(x,y, '*-')
plt.xlabel("Tid i timmar")
plt.ylabel("Antal bakterier")
plt.title("Bakterier i mjölk i rumstemperatur")


```
